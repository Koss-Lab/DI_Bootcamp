```python
import zipfile, os
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
from scipy.ndimage import rotate

```

```python
import zipfile
import os
from google.colab import files

# Ask user to upload the dataset
print("📂 Please upload the Flower Color Images Dataset.zip file")
uploaded = files.upload()

# Get the uploaded file name dynamically
zip_path = list(uploaded.keys())[0]
extract_path = "/content/flowers"

# Extract dataset
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("✅ Dataset extracted to:", extract_path)
print("📂 Subfolders:", os.listdir(extract_path))

```

```python
# List available files to find the correct image path
import glob

# Search inside extracted dataset
image_files = glob.glob(os.path.join(extract_path, "**/*.png"), recursive=True)
print("🔎 Found", len(image_files), "images")
print("Example path:", image_files[0] if len(image_files) > 0 else "❌ No images found")

# Load first image if exists
if len(image_files) > 0:
    image_path = image_files[0]
    original_image = Image.open(image_path)

    plt.imshow(original_image)
    plt.title("Original Image")
    plt.axis("off")
    plt.show()
else:
    print("⚠️ Please upload the dataset or check the extraction folder.")
```

```python
def rotate_image(image, angle=90):
    return rotate(image, angle, reshape=False)

rotated_img = rotate_image(original_image, 90)

plt.imshow(rotated_img)
plt.title("Rotated 90°")
plt.axis("off")
plt.show()
```

```python
# Horizontal flip
flipped_h = ImageOps.mirror(original_image)

# Vertical flip
flipped_v = ImageOps.flip(original_image)

# Display both
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
axes[0].imshow(flipped_h)
axes[0].set_title("Horizontal Flip")
axes[0].axis("off")

axes[1].imshow(flipped_v)
axes[1].set_title("Vertical Flip")
axes[1].axis("off")

plt.show()
```

```python
def zoom_image(image, scale=1.2):
    w, h = image.size
    return image.resize((int(w*scale), int(h*scale)))

zoomed_img = zoom_image(original_image, 1.2)

plt.imshow(zoomed_img)
plt.title("Zoomed Image (1.2x)")
plt.axis("off")
plt.show()
```
