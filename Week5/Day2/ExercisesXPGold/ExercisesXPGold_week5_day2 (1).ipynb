```python
# Exercise 1
import pandas as pd
from google.colab import files

# Upload the three retail datasets
uploaded = files.upload()

# Load them (make sure the filenames match exactly what you upload)
sales = pd.read_csv("retaildata_sales_data-set.csv")
features = pd.read_csv("retaildata_Features_data_set.csv")
stores = pd.read_csv("retaildata_stores_data-set.csv")

# Merge into one dataframe
retail_df = sales.merge(features, on=["Store", "Date"], how="left").merge(stores, on="Store", how="left")

print(retail_df.head())
print(retail_df.info())

# Upload and load the clothing reviews dataset
uploaded = files.upload()
reviews = pd.read_csv("Womens_Clothing E-Commerce_Reviews.csv")

print(reviews.head())
print(reviews.info())
```
