```python
# Exercise 1

import matplotlib.pyplot as plt

# Fictional dataset
categories = ["Electronics", "Clothing", "Home", "Books", "Toys"]
sales = [25000, 18000, 22000, 12000, 15000]

# Bar graph
plt.figure(figsize=(8,6))
plt.bar(categories, sales, color="teal")
plt.xlabel("Product Category")
plt.ylabel("Total Sales ($)")
plt.title("Retail Company Sales by Category")
plt.show()
```

```python
# Exercise 2

import pandas as pd

# Sample data
data = {
    ("Canada", "Toronto", "2025-07-01"): 32,
    ("Canada", "Toronto", "2025-07-02"): 29,
    ("Canada", "Vancouver", "2025-07-01"): 28,
    ("Canada", "Vancouver", "2025-07-02"): 31,
    ("USA", "New York", "2025-07-01"): 35,
}

# Create hierarchical DataFrame
df = pd.Series(data, name="Temperature").to_frame()
df.index = pd.MultiIndex.from_tuples(df.index, names=["Country", "City", "Date"])

# Filter: Canada + temp > 30
filtered = df.loc["Canada"][df.loc["Canada"]["Temperature"] > 30]
print(filtered)
```

```python
# Exercise 3

# Example hierarchical dataset (Department → Employee → Salary)
data = {
    ("IT", "Alice"): 60000,
    ("IT", "Bob"): 48000,
    ("HR", "Carol"): 52000,
    ("HR", "Dan"): 45000,
    ("Finance", "Eve"): 70000,
}
df = pd.Series(data, name="Salary").to_frame()
df.index = pd.MultiIndex.from_tuples(df.index, names=["Department", "Employee"])

# Filter: Salary > 50,000 within each department
filtered = df[df["Salary"] > 50000]
print(filtered)
```

```python
# Exercise 4

import seaborn as sns
import matplotlib.pyplot as plt

# Example MCU durations dataset
mcu_data = {
    "Movie": ["Iron Man", "Avengers", "Doctor Strange", "Black Panther", "Endgame"],
    "Duration": [126, 143, 115, 134, 181]
}
df_movies = pd.DataFrame(mcu_data)

# Distribution plot
plt.figure(figsize=(8,6))
sns.histplot(df_movies["Duration"], kde=True, color="purple", bins=8)
plt.title("Distribution of MCU Movie Durations")
plt.xlabel("Duration (minutes)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Exercise 5

# Example dataset
data = {
    "Movie": ["Iron Man", "Avengers", "Doctor Strange", "Black Panther", "Endgame"],
    "Tomato Meter": [94, 91, 89, 96, 94],
    "Audience Score": [91, 92, 86, 79, 94]
}
df_scores = pd.DataFrame(data)

# Melt to long format
df_long = df_scores.melt(id_vars="Movie", value_vars=["Tomato Meter", "Audience Score"],
                         var_name="Score Type", value_name="Score")

# Box plot
plt.figure(figsize=(8,6))
sns.boxplot(x="Score Type", y="Score", data=df_long, palette="Set2")
plt.title("Distribution of Tomato Meter vs Audience Scores")
plt.show()
```

```python
# Exercise 6

# Example financial dataset
financial_data = {
    "Movie": ["Iron Man", "Avengers", "Doctor Strange", "Black Panther", "Endgame"],
    "Production Budget": [140000000, 220000000, 165000000, 200000000, 356000000],
    "Opening Weekend": [102000000, 207000000, 85000000, 202000000, 357000000],
    "Domestic Box Office": [318000000, 623000000, 232000000, 700000000, 858000000],
    "Worldwide Box Office": [585000000, 1518000000, 677000000, 1347000000, 2797000000],
    "MCU Phase": ["Phase 1", "Phase 1", "Phase 3", "Phase 3", "Phase 3"]
}
df_fin = pd.DataFrame(financial_data)

# Pair plot
sns.pairplot(df_fin,
             vars=["Production Budget", "Opening Weekend", "Domestic Box Office", "Worldwide Box Office"],
             hue="MCU Phase", palette="husl")
plt.show()
```
