Data visualization is important in data analysis because it helps to transform complex data into an understandable and interpretable format. It allows analysts to quickly identify trends, patterns, and outliers that may not be evident in raw data tables. Visualizations also make it easier to communicate insights effectively to non-technical stakeholders.

The purpose of a line graph in data visualization is to show how a variable changes over time. It is particularly useful for displaying time-series data such as temperature variations, stock prices, or sales trends, allowing us to observe trends, fluctuations, and patterns across continuous intervals.


```python
# Exercise 2

import matplotlib.pyplot as plt

# Data
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
temperatures = [72, 74, 76, 80, 82, 78, 75]

# Plot
plt.plot(days, temperatures, marker="o", linestyle="-", color="blue")
plt.xlabel("Day")
plt.ylabel("Temperature (Â°F)")
plt.title("Temperature Variation Over a Week")
plt.show()
```

```python
# Exercise 3

# Data
months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [5000, 5500, 6200, 7000, 7500]

# Plot
plt.bar(months, sales, color="green")
plt.xlabel("Month")
plt.ylabel("Sales Amount ($)")
plt.title("Monthly Sales Data")
plt.show()
```

```python
# Exercise 4

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Ask user to upload the file in Colab
uploaded = files.upload()

# Load dataset after upload
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

# Histogram of CGPA
plt.figure(figsize=(8,5))
sns.histplot(df["What is your CGPA?"], color="purple")
plt.title("Distribution of Students' CGPA")
plt.xlabel("CGPA Range")
plt.ylabel("Frequency")
plt.show()
```

```python
# Exercise 5

plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Choose your gender", hue="Do you have Anxiety?", palette="Set2")
plt.title("Anxiety Levels Across Genders")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
# Exercise 6

# Convert Panic Attack responses to numeric (Yes=1, No=0)
df["PanicAttackNumeric"] = df["Do you have Panic attack?"].map({"Yes": 1, "No": 0})

plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="Age", y="PanicAttackNumeric", hue="PanicAttackNumeric", palette="coolwarm")
plt.title("Relationship Between Age and Panic Attacks")
plt.xlabel("Age")
plt.ylabel("Panic Attack (1=Yes, 0=No)")
plt.show()
```
